```python
cat=[col for col in df.columns if df[col].nunique()<6]
num=[col for col in df.columns if df[col].nunique()>6]
```

```python
print(cat)
print(num)
```

```python
std=StandardScaler()
column=ColumnTransformer(
    [
        ("numerical_transformation",std,num)
    ]
)
```

```python
X=df.drop('Outcome',axis=1)
Y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=42,test_size=0.27)
```

```python
x_test
```

```python
x_train=column.fit_transform(x_train)
x_test=column.transform(x_test)
```

```python
x_train
```

```python
import pickle
import dill

with open('preprocessor.pkl','wb') as file:
    dill.dump(column,file)
```

```python
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
model=Sequential(
    [
        Dense(64,activation=('relu'),input_shape=(x_train.shape[1],)),
        Dense(32,activation=('relu')),  #Hidden Layer
        Dense(1,activation='sigmoid') #Output layer
    ]
)
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
```

```python
model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

setup Tensorboard

```python

logs='log/fit'+ datetime.datetime.now().strftime("%Y%M%d%H%M%S")
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
ten=TensorBoard(log_dir=logs,histogram_freq=1)
early=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[ten,early]
)
```

```python
model.save('model.h5')
```

```python
from scikeras.wrappers import KerasClassifier
```

```python
def model_tunning(neurons=30,layers=1):
    model1=Sequential()

    model1.add(Dense(neurons,activation='relu',input_shape=(x_train.shape[1],)))

    for _ in range(layers-1):
        model1.add(neurons,activation='relu')

    
    model1.add(Dense(1,activation='sigmoid'))
    
    model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy']) 

    return model1   
```

```python
model1=KerasClassifier(build_fn=model_tunning,epochs=50,verbose=0,batch_size=10,layers=1,neurons=30)
```

```python
param_grid={
    "neurons":[30,40,50,60],
    "layers":[1,3,4,5],
    "epochs":[50,60,70]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(estimator=model1,param_grid=param_grid,cv=5)
grid.fit(x_train,y_train)

print(grid.best_params_,grid.best_score_)
```
